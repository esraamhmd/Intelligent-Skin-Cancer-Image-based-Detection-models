```python
import os
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, models
from PIL import Image
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    accuracy_score, classification_report, confusion_matrix, 
    roc_curve, auc, precision_recall_curve, average_precision_score
)
from sklearn.preprocessing import label_binarize, StandardScaler
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from tqdm import tqdm
import time
import random
import joblib
import ssl
import gc  
import multiprocessing
from concurrent.futures import ThreadPoolExecutor, ProcessPoolExecutor
import datetime

# Fix for SSL certificate issue
ssl._create_default_https_context = ssl._create_unverified_context

# Function to log messages 
def log_message(message, log_file="/kaggle/working/progress_log.txt"):
    timestamp = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    formatted_message = f"[{timestamp}] {message}"
    
    # Print to console
    print(formatted_message)
    
    # Write to log file
    with open(log_file, "a") as f:
        f.write(formatted_message + "\n")


cpu_count = multiprocessing.cpu_count()
optimal_threads = max(1, cpu_count - 1)  

# Set environment variables for optimized performance
os.environ['OMP_NUM_THREADS'] = str(optimal_threads)
os.environ['MKL_NUM_THREADS'] = str(optimal_threads)
os.environ['NUMEXPR_NUM_THREADS'] = str(optimal_threads)
os.environ['VECLIB_MAXIMUM_THREADS'] = str(optimal_threads)
os.environ['OPENBLAS_NUM_THREADS'] = str(optimal_threads) 

# Function to setup device (TPU/GPU/CPU)
def setup_device():
    try:
        import torch_xla
        import torch_xla.core.xla_model as xm
        device = xm.xla_device()
        log_message("Using TPU for computation")
        return device, "tpu"
    except ImportError:
        if torch.cuda.is_available():
            torch.cuda.set_device(0)
            log_message(f"Using GPU: {torch.cuda.get_device_name(0)}")
            return torch.device("cuda"), "cuda"
        else:
            log_message("Using CPU for computation")
            return torch.device("cpu"), "cpu"

# Set up device
device, device_type = setup_device()

# Optimize CUDA operations if using GPU
if device_type == 'cuda':
    # Optimize CUDA operations
    torch.backends.cudnn.benchmark = True
    torch.backends.cudnn.deterministic = False
    torch.backends.cuda.matmul.allow_tf32 = True
    torch.backends.cudnn.allow_tf32 = True
    
    # Clear CUDA cache
    torch.cuda.empty_cache()
    
    # Monitor memory
    torch.cuda.memory.memory_stats() 
    torch.cuda.memory.reset_peak_memory_stats()

# Use mixed precision training 
if device_type == 'cuda' and torch.cuda.get_device_capability()[0] >= 7:
    log_message("Enabling mixed precision training")
    torch.backends.cuda.matmul.allow_fp16_reduced_precision_reduction = True
    
    from torch.cuda.amp import autocast, GradScaler
    use_amp = True
else:
    use_amp = False

# Memory management
def release_memory(verbose=False):
    """Release unused memory in PyTorch and trigger garbage collection"""
    before = 0
    if torch.cuda.is_available() and verbose:
        before = torch.cuda.memory_allocated()
        
    if torch.cuda.is_available():
        torch.cuda.empty_cache()
        
    collected = gc.collect(generation=2)  
    
    if torch.cuda.is_available() and verbose:
        after = torch.cuda.memory_allocated()
        log_message(f"Memory freed: {(before-after)/1024**2:.2f} MB, Objects collected: {collected}")

# Initialize progress update timer
last_progress_update = time.time()

def check_progress_update(message):
    """Check if 10 minutes have passed to update progress"""
    global last_progress_update
    current_time = time.time()
    
    if current_time - last_progress_update >= 600:  # 10 minutes = 600 seconds
        log_message(f"Progress update: {message}")
        last_progress_update = current_time
        return True
    return False

plt.ioff()  
plt.rcParams['figure.dpi'] = 100  
plt.style.use('fast') 

# Optimize NumPy operations
np.seterr(divide='ignore', invalid='ignore') 
np.set_printoptions(precision=4, suppress=True)  

# Set random seeds 
torch.manual_seed(42)
random.seed(42)
np.random.seed(42)

# Define target counts for binary classification (malignant vs benign)
target_counts = {
    'malignant': 8388,  # Combined target for mel and bcc
    'benign': 8388     # Combined target for akiec, bkl, df, nv, vasc
}

# Dictionary to store model paths
model_paths = {
    'svm': "/kaggle/working/feature_fusion_svm_binary_ham10000.joblib",
    'random_forest': "/kaggle/working/feature_fusion_rf_binary_ham10000.joblib"
}

# Dictionaries to store metrics
model_train_times = {}
model_eval_times = {}
model_total_times = {}
all_metrics = {}

# DATA PROCESSING FUNCTIONS 

def augment_and_prepare_dataset(dataset_path):
    log_message("\nLoading and augmenting dataset...")
    
    # Load metadata 
    metadata_file = os.path.join(dataset_path, "HAM10000_metadata.csv")
    metadata = pd.read_csv(metadata_file, low_memory=False)
    
    # Create binary labels for malignant vs benign
    # 'mel' and 'bcc' are malignant (1), rest  is benign (0)
    metadata['binary_label'] = metadata['dx'].apply(lambda x: 'malignant' if x in ['mel', 'bcc'] else 'benign')
    
    # Calculate initial class distribution
    initial_class_counts = metadata['binary_label'].value_counts()
    log_message("\nInitial binary class distribution:")
    for cls, count in initial_class_counts.items():
        log_message(f"{cls}: {count}")
    
    # Get image paths
    image_folders = ["HAM10000_images_part_1", "HAM10000_images_part_2"]
    image_path_dict = {}
    
    check_progress_update("Processing image folders...")
    
    def process_folder(folder):
        local_dict = {}
        folder_path = os.path.join(dataset_path, folder)
        if os.path.exists(folder_path):
            for img_name in os.listdir(folder_path):
                if img_name.endswith('.jpg'):
                    image_id = os.path.splitext(img_name)[0]
                    local_dict[image_id] = os.path.join(folder_path, img_name)
        return local_dict
    
    # Process folders in parallel 
    if len(image_folders) > 1:
        with ThreadPoolExecutor(max_workers=len(image_folders)) as executor:
            results = list(executor.map(process_folder, image_folders))
            
        for result in results:
            image_path_dict.update(result)
    else:
        image_path_dict = process_folder(image_folders[0])
    
    # Add image paths to metadata 
    metadata['image_path'] = metadata['image_id'].map(image_path_dict)
    
    # Create a plot to visualize initial class distribution
    plt.figure(figsize=(12, 6))
    ax = sns.barplot(x=initial_class_counts.index, y=initial_class_counts.values)
    plt.title('Initial Binary Class Distribution', fontsize=15)
    plt.xlabel('Class', fontsize=12)
    plt.ylabel('Number of Images', fontsize=12)
    
    # Add count labels above the bars
    for i, count in enumerate(initial_class_counts.values):
        ax.text(i, count + 50, str(count), ha='center', fontsize=10, fontweight='bold')
    
    plt.subplots_adjust(bottom=0.15, top=0.9)
    plt.savefig("/kaggle/working/initial_binary_class_distribution.png", dpi=300)
    plt.close()  
    
    check_progress_update("Balancing classes...")
    
    # Balancing classes 
    processed_metadata = []
    
    # For each binary class, downsample if needed
    for class_name in metadata['binary_label'].unique():
        class_samples = metadata[metadata['binary_label'] == class_name]
        target_count = target_counts[class_name]
        
        if len(class_samples) > target_count:
            # Downsample class to target count
            log_message(f"Downsampling '{class_name}' class from {len(class_samples)} to {target_count} samples")
            class_samples = class_samples.sample(target_count, random_state=42)
        
        processed_metadata.append(class_samples)
    
    # Combine all processed classes
    processed_metadata = pd.concat(processed_metadata, ignore_index=True)
    
    # Calculate new class distribution after downsampling
    processed_class_counts = processed_metadata['binary_label'].value_counts()
    log_message("\nClass distribution after downsampling:")
    for cls, count in processed_class_counts.items():
        log_message(f"{cls}: {count}")
    
    # Define augmentation transforms 
    augmentation_transforms = [
        transforms.RandomRotation(15),  # 15 degree rotation
        transforms.RandomHorizontalFlip(p=0.5),  # Horizontal flip
        transforms.RandomVerticalFlip(p=0.5),  # Vertical flip
        transforms.RandomAffine(
            degrees=0,  # No additional rotation
            translate=(0.1, 0.1),  # Width and height shift of 10%
            scale=(0.8, 1.2),  # Zoom effect with scale range of 0.2 (20%)
        ),
    ]
    
    # Use the same augmentations but at different probabilities for diversity
    strong_augmentation_transforms = [
        transforms.RandomRotation(15),  # 15 degree rotation
        transforms.RandomHorizontalFlip(p=0.8),  # Horizontal flip 
        transforms.RandomVerticalFlip(p=0.8),  # Vertical flip 
        transforms.RandomAffine(
            degrees=0,  # No additional rotation
            translate=(0.1, 0.1),  # Width and height shift of 10%
            scale=(0.8, 1.2),  # Zoom effect with scale range of 0.2 (20%)
        ),
    ]
    
    # STEP 1: Create all images (original + augmented) as PIL Images
    log_message("\nSTEP 1: Processing original images and generating augmentations...")
    all_images = []  
    
    batch_size = 128  # Process images in batches
    
    # Create a function to process a batch of images
    def process_image_batch(batch_rows):
        batch_results = []
        for _, row in batch_rows.iterrows():
            try:
                img_path = row['image_path']
                label = row['binary_label']  # Use binary label
                image_id = row['image_id']
                
                if pd.isna(img_path):
                    continue
                    
                # Load original image and add to the list
                img = Image.open(img_path).convert('RGB')
                batch_results.append((img, label, f"{image_id}_orig"))
                
            except Exception as e:
                log_message(f"Error processing image {row['image_id']}: {e}")
        return batch_results
    
    # Process original images in batches
    num_batches = len(processed_metadata) // batch_size + (1 if len(processed_metadata) % batch_size > 0 else 0)
    
    for i in tqdm(range(num_batches), desc="Processing original images"):
        check_progress_update(f"Processing original images - batch {i+1}/{num_batches}")
        
        start_idx = i * batch_size
        end_idx = min((i + 1) * batch_size, len(processed_metadata))
        batch = processed_metadata.iloc[start_idx:end_idx]
        all_images.extend(process_image_batch(batch))
        
        # Release memory 
        if i % 10 == 0:
            release_memory()
    
    # Create augmented data
    log_message("\nGenerating augmented images...")
    
    for class_name, target_count in target_counts.items():
        # Get original samples for this class
        class_samples = processed_metadata[processed_metadata['binary_label'] == class_name]
        current_count = len(class_samples)
        
        check_progress_update(f"Processing class {class_name} - current count: {current_count}, target: {target_count}")
        
        if current_count >= target_count:
            log_message(f"Class '{class_name}' already has {current_count} samples, no augmentation needed")
            continue
            
        # Calculate how many augmented samples we need
        num_augmented_needed = target_count - current_count
        
        log_message(f"Augmenting '{class_name}' class: adding {num_augmented_needed} samples to reach {target_count}")
        
        # Generate augmented samples 
        num_per_original = max(1, num_augmented_needed // len(class_samples) + 1)
        counter = 0
        
        # Pre-load images 
        preloaded_images = {}
        for _, row in tqdm(class_samples.iterrows(), desc=f"Preloading images for {class_name}", total=len(class_samples)):
            try:
                img_path = row['image_path']
                if pd.isna(img_path):
                    continue
                preloaded_images[row['image_id']] = Image.open(img_path).convert('RGB')
            except Exception as e:
                log_message(f"Error preloading image {row['image_id']}: {e}")
        
        augmentation_count = 0
        indices = list(range(len(class_samples)))
        
        while counter < num_augmented_needed:
            # Check for progress update
            check_progress_update(f"Augmenting class {class_name}: {counter}/{num_augmented_needed} samples generated")
            
            # Cycle through the indices
            idx = indices[augmentation_count % len(indices)]
            sample = class_samples.iloc[idx]
            image_id = sample['image_id']
            
            try:
                # Get the preloaded image
                if image_id not in preloaded_images:
                    augmentation_count += 1
                    continue
                    
                img = preloaded_images[image_id]
                
                # Decide which augmentation strategy to use
                if counter % 5 == 0:  # Every 5th augmentation, use strong mix
                    # Apply the strong augmentations
                    aug_img = img
                    for transform in strong_augmentation_transforms:
                        aug_img = transform(aug_img)
                
                else:
                    # Apply the standard augmentations
                    aug_img = img
                    for transform in augmentation_transforms:
                        aug_img = transform(aug_img)
                
                # Add augmented image to the list
                all_images.append((aug_img, class_name, f"{sample['image_id']}_aug_{counter}"))
                counter += 1
                
                if counter % 3000 == 0:
                    log_message(f"Generated {counter}/{num_augmented_needed} augmented samples for class {class_name}")
                    release_memory()
                    
            except Exception as e:
                log_message(f"Error generating augmented image: {e}")
            
            augmentation_count += 1
    
    # Clean up preloaded images to free memory
    preloaded_images = None
    release_memory()
    
    # STEP 2: Create transforms 
    log_message(f"\nSTEP 2: Creating transforms ...")
    
    check_progress_update("Creating image transforms...")
    
    # Define transform for all models
    transforms_dict = {
        'mobilenet': transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ]),
        'densenet121': transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ]),
        'resnet50': transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
    }
    # Process all images with the common transform for dataset splitting
    all_processed_data = []
    # Use MobileNetV3 transforms as default for splitting purposes
    default_transform = transforms_dict['mobilenet']
    
    log_message(f"Applying resize and normalize for dataset splitting...")
    for i, (img, label, image_id) in enumerate(tqdm(all_images, desc="Processing images")):
        # Check for progress updates
        if i % 100 == 0:
            check_progress_update(f"Processing images: {i}/{len(all_images)}")
            
        img_tensor = default_transform(img)
        all_processed_data.append((img_tensor, label, image_id, img))
        
    # STEP 3: Split the dataset
    log_message("\nSTEP 3: Splitting dataset into train, validation, and test sets...")
    
    check_progress_update("Splitting dataset...")
    
    # Create class to index mapping for binary classification
    classes = sorted(list(set([label for _, label, _ in all_images])))
    class_to_idx = {cls: i for i, cls in enumerate(classes)}
    log_message(f"Binary class mapping: {class_to_idx}")
    
    # Convert string labels to indices
    processed_data = [(img, class_to_idx[label], image_id) for img, label, image_id in all_images]
    
    # Determine the exact split sizes 
    total_size = len(processed_data)
    train_size = int(0.70 * total_size)
    remaining = total_size - train_size
    val_size = remaining // 2  
    test_size = remaining - val_size  
    
    if val_size != test_size:
        train_size += 1
        val_size = test_size  # Now they're exactly equal
        
    log_message(f"Splitting dataset: {train_size} train, {val_size} validation, {test_size} test samples")
    
    # Stratified splitting to maintain class distribution
    random.seed(42)
    random.shuffle(processed_data)
    
    # Extract labels 
    labels = [label for _, label, _ in processed_data]
    
    indices = np.arange(total_size)
    train_indices, temp_indices = train_test_split(
        indices, test_size=val_size+test_size, 
        stratify=labels, 
        random_state=42
    )
    
    val_indices, test_indices = train_test_split(
        temp_indices, 
        test_size=0.5,  # Exactly half for each
        stratify=[labels[i] for i in temp_indices],
        random_state=42
    )
    
    # Create datasets
    train_data = [processed_data[i] for i in train_indices]
    val_data = [processed_data[i] for i in val_indices]
    test_data = [processed_data[i] for i in test_indices]
    
    check_progress_update("Creating datasets...")
    
    # Create dataset class for feature extraction
    class SkinLesionDataset(Dataset):
        def __init__(self, data, transform=None):
            self.data = data
            self.transform = transform
            
        def __len__(self):
            return len(self.data)
            
        def __getitem__(self, idx):
            img, label, _ = self.data[idx]
            if self.transform:
                img = self.transform(img)
            return img, label
    
    # Optimize batch_size and num_workers based on system resources
    if device_type == 'tpu':
        batch_size = 128
        num_workers = 96
    elif device_type == 'cuda':
        gpu_memory = torch.cuda.get_device_properties(0).total_memory / 1e9
        
        if gpu_memory > 16:  
            batch_size = 512
        elif gpu_memory > 8:  
            batch_size = 256
        else:  
            batch_size = 128
        
        num_workers = min(96, os.cpu_count() or 4)
    else:
        # Default batch size for CPU
        batch_size = 128
        num_workers = min(128, os.cpu_count() or 4)
    
    log_message(f"Using batch_size={batch_size}, num_workers={num_workers}")
    
    # Create optimized data loaders 
    train_dataset = SkinLesionDataset(train_data, transform=default_transform)
    val_dataset = SkinLesionDataset(val_data, transform=default_transform)
    test_dataset = SkinLesionDataset(test_data, transform=default_transform)
    
    train_loader = DataLoader(
        train_dataset, 
        batch_size=batch_size, 
        shuffle=False, 
        num_workers=num_workers, 
        pin_memory=(device_type != 'tpu'),
        persistent_workers=(num_workers > 0 and device_type != 'tpu'),
        prefetch_factor=2 if num_workers > 0 else None
    )
    
    val_loader = DataLoader(
        val_dataset, 
        batch_size=batch_size, 
        shuffle=False, 
        num_workers=num_workers, 
        pin_memory=(device_type != 'tpu'),
        persistent_workers=(num_workers > 0 and device_type != 'tpu'),
        prefetch_factor=2 if num_workers > 0 else None
    )
    
    test_loader = DataLoader(
        test_dataset, 
        batch_size=batch_size, 
        shuffle=False, 
        num_workers=num_workers, 
        pin_memory=(device_type != 'tpu'),
        persistent_workers=(num_workers > 0 and device_type != 'tpu'),
        prefetch_factor=2 if num_workers > 0 else None
    )
    
    # Count samples in each split by class
    def count_classes(data_list):
        count_dict = {}
        for _, label, _ in data_list:
            if label not in count_dict:
                count_dict[label] = 0
            count_dict[label] += 1
        return count_dict
    
    train_counts = count_classes(train_data)
    val_counts = count_classes(val_data)
    test_counts = count_classes(test_data)
    
    # Convert class indices back to names for display
    idx_to_class = {v: k for k, v in class_to_idx.items()}
    
    # Display class distribution in each split
    log_message("\nClass distribution in train split:")
    for label, count in train_counts.items():
        log_message(f"  Class {idx_to_class[label]}: {count} images")
    
    log_message("\nClass distribution in validation split:")
    for label, count in val_counts.items():
        log_message(f"  Class {idx_to_class[label]}: {count} images")
    
    log_message("\nClass distribution in test split:")
    for label, count in test_counts.items():
        log_message(f"  Class {idx_to_class[label]}: {count} images")
    
    # Clear processed data to free memory
    processed_data = None
    all_images = None
    release_memory()
    
    return {
        'train_loader': train_loader,
        'val_loader': val_loader,
        'test_loader': test_loader,
        'class_to_idx': class_to_idx,
        'idx_to_class': idx_to_class,
        'classes': classes,
        'num_classes': len(classes)
    }

class MobileNetV3Extractor(nn.Module):
    """Feature extractor using MobileNetV3 without the classifier"""
    def __init__(self):
        super(MobileNetV3Extractor, self).__init__()
      
        model = models.mobilenet_v3_large(weights=models.MobileNet_V3_Large_Weights.IMAGENET1K_V1)
        self.features = nn.Sequential(*list(model.children())[:-1])
        
    def forward(self, x):
        x = self.features(x)
        return x.squeeze()

class DenseNet121Extractor(nn.Module):
    """Feature extractor using DenseNet121 without the classifier"""
    def __init__(self):
        super(DenseNet121Extractor, self).__init__()
       
        model = models.densenet121(weights=models.DenseNet121_Weights.IMAGENET1K_V1)
        self.features = nn.Sequential(*list(model.children())[:-1])
        
    def forward(self, x):
        x = self.features(x)
        # Apply global average pooling
        x = torch.nn.functional.adaptive_avg_pool2d(x, (1, 1))
        return x.squeeze()

class ResNet50Extractor(nn.Module):
    """Feature extractor using ResNet50 without the classifier"""
    def __init__(self):
        super(ResNet50Extractor, self).__init__()
        
        model = models.resnet50(weights=models.ResNet50_Weights.IMAGENET1K_V1)
        self.features = nn.Sequential(*list(model.children())[:-1])
        
    def forward(self, x):
        x = self.features(x)
        return x.squeeze()

# FEATURE EXTRACTION FUNCTIONS 

def extract_features_from_model(model, data_loader, device, device_type, model_name):
    """Extract features using a specific model with optimized batch processing"""
    model.to(device)
    model.eval()
    
    features = []
    batch_count = 0
    
    with torch.no_grad():
        for i, (images, _) in enumerate(tqdm(data_loader, desc=f"Extracting {model_name} features")):
            # Check progress update
            if i % 5 == 0:
                check_progress_update(f"Extracting {model_name} features: batch {i}/{len(data_loader)}")
                
            # Handle TPU differently
            if device_type == 'tpu':
                import torch_xla.core.xla_model as xm
                images = images.to(device)
                batch_features = model(images)
                # Move features to CPU
                batch_features = xm.all_gather(batch_features).cpu().numpy()
            else:
                images = images.to(device, non_blocking=True)
                
                if use_amp and device_type == 'cuda':
                    with autocast():
                        batch_features = model(images)
                else:
                    batch_features = model(images)
                    
                batch_features = batch_features.cpu().numpy()
                
            features.append(batch_features)
            
            batch_count += 1
            if batch_count % 20 == 0: 
                if device_type == 'cuda':
                    torch.cuda.empty_cache()
    
    # stack all features
    return np.vstack(features)

# extract features
def extract_features_and_labels(data_loader, device, device_type):
    """Extract features from all models separately and combine with optimized memory handling"""
   
    log_message("\nInitializing feature extractors...")
    mobilenet_extractor = MobileNetV3Extractor()
    
    # Extract labels  
    labels = []
    for _, targets in data_loader:
        if device_type == 'tpu':
            labels.append(targets.cpu().numpy())
        else:
            labels.append(targets.numpy())
    labels = np.concatenate(labels)
    
    # Extract features from each model separately
    log_message("\nExtracting features from MobileNetV3...")
    mobilenet_features = extract_features_from_model(mobilenet_extractor, data_loader, device, device_type, "MobileNetV3")
    log_message(f"MobileNetV3 features shape: {mobilenet_features.shape}")
    
    # Clear the first model to free up GPU memory
    mobilenet_extractor = None
    release_memory()
    
    check_progress_update("Starting DenseNet121 feature extraction")
    
    log_message("\nExtracting features from DenseNet121...")
    densenet_extractor = DenseNet121Extractor()
    densenet_features = extract_features_from_model(densenet_extractor, data_loader, device, device_type, "DenseNet121")
    log_message(f"DenseNet121 features shape: {densenet_features.shape}")
    
    # Clear the second model to free up GPU memory
    densenet_extractor = None
    release_memory()
    
    check_progress_update("Starting ResNet50 feature extraction")
    
    log_message("\nExtracting features from ResNet50...")
    resnet50_extractor = ResNet50Extractor()
    resnet50_features = extract_features_from_model(resnet50_extractor, data_loader, device, device_type, "ResNet50")
    log_message(f"ResNet50 features shape: {resnet50_features.shape}")
    
    # Clear the third model to free up GPU memory
    resnet50_extractor = None
    release_memory()
    
    feature_dict = {
        'mobilenet': mobilenet_features,
        'densenet': densenet_features,
        'resnet50': resnet50_features
    }
    
    # create fused features by concatenating
    fused_features = np.concatenate([
        mobilenet_features, 
        densenet_features, 
        resnet50_features
    ], axis=1)
    
    log_message(f"\nFused features shape: {fused_features.shape}")
    
    # Print label distribution
    unique_labels, counts = np.unique(labels, return_counts=True)
    log_message("\nLabel distribution in this split:")
    for label, count in zip(unique_labels, counts):
        log_message(f"  Class {label}: {count} images")
    
    # Convert to float32 to reduce memory usage
    for key in feature_dict:
        feature_dict[key] = feature_dict[key].astype(np.float32)
    
    fused_features = fused_features.astype(np.float32)
    
    return feature_dict, fused_features, labels

# MODEL TRAINING FUNCTIONS 

def train_individual_models(feature_dict, y_train, y_val, X_val_dict):
    """Train individual models with features from each feature extractor with enhanced parameters"""
    results = {}
    
    # Scale the features 
    scalers = {}
    scaled_features = {}
    scaled_val_features = {}
    
    # Scale each feature set separately
    for model_name, features in feature_dict.items():
        log_message(f"Scaling features for {model_name}...")
        scaler = StandardScaler()
        scaled_features[model_name] = scaler.fit_transform(features)
        scaled_val_features[model_name] = scaler.transform(X_val_dict[model_name])
        scalers[model_name] = scaler
    
    # SVM parameters 
    SVM_PARAMS = {
        'kernel': 'rbf',
        'probability': True,
        'C': 1.0,
        'gamma': 'scale',  
        'cache_size': 3000,  
        'max_iter': 400,   
        'tol': 1e-2,        
        'class_weight': None,  
        'verbose': False,    
        'decision_function_shape': 'ovo',  
        'break_ties': False  
     }
     
    # RF parameters 
    RF_PARAMS = {
        'n_estimators': 100,     
        'random_state': 42,
        'n_jobs': -1,            
        'max_depth': 10,        
        'min_samples_split': 10, 
        'min_samples_leaf': 4,   
        'bootstrap': True,      
        'verbose': 0,           
        'warm_start': False,
        'max_features': 'sqrt',  
        'criterion': 'gini',     
        'max_leaf_nodes': 100,
        'class_weight': None  
    }
    
    for model_name, features in scaled_features.items():
        log_message(f"\n{'='*50}")
        log_message(f"Training SVM using {model_name} features...")
        
        check_progress_update(f"Training SVM with {model_name} features")
        
        # Record start time
        start_time = time.time()
        
        # Train SVM with optimized parameters
        svm_classifier = SVC(**SVM_PARAMS)
        svm_classifier.fit(features, y_train)
        
        # Calculate training time
        train_time = time.time() - start_time
        model_train_times[f'{model_name}_svm'] = train_time
        
        # Training metrics
        y_train_pred = svm_classifier.predict(features)
        train_accuracy = accuracy_score(y_train, y_train_pred)
        
        # Validation metrics
        val_features = scaled_val_features[model_name]
        y_val_pred = svm_classifier.predict(val_features)
        val_accuracy = accuracy_score(y_val, y_val_pred)
        
        log_message(f"{model_name} SVM Training Time: {train_time:.2f} seconds")
        log_message(f"{model_name} SVM Training Accuracy: {train_accuracy:.4f}")
        log_message(f"{model_name} SVM Validation Accuracy: {val_accuracy:.4f}")
        
        # Save model and scaler
        joblib.dump(svm_classifier, f"/kaggle/working/{model_name}_svm_binary_ham10000.joblib")
        joblib.dump(scalers[model_name], f"/kaggle/working/{model_name}_scaler_svm_binary_ham10000.joblib")
        
        # Store results
        results[f'{model_name}_svm'] = {
            'model': svm_classifier,
            'scaler': scalers[model_name],
            'train_acc': train_accuracy,
            'val_acc': val_accuracy
        }
        
        # Train Random Forest
        log_message(f"\n{'='*50}")
        log_message(f"Training Random Forest using {model_name} features...")
        
        check_progress_update(f"Training Random Forest with {model_name} features")
        
        # Record start time
        start_time = time.time()
        
        rf_classifier = RandomForestClassifier(**RF_PARAMS)
        rf_classifier.fit(features, y_train)
        
        # Calculate training time
        train_time = time.time() - start_time
        model_train_times[f'{model_name}_rf'] = train_time
        
        # Training metrics
        y_train_pred = rf_classifier.predict(features)
        train_accuracy = accuracy_score(y_train, y_train_pred)
        
        # Validation metrics
        y_val_pred = rf_classifier.predict(val_features)
        val_accuracy = accuracy_score(y_val, y_val_pred)
        
        log_message(f"{model_name} RF Training Time: {train_time:.2f} seconds")
        log_message(f"{model_name} RF Training Accuracy: {train_accuracy:.4f}")
        log_message(f"{model_name} RF Validation Accuracy: {val_accuracy:.4f}")
        
        # Save model
        joblib.dump(rf_classifier, f"/kaggle/working/{model_name}_rf_binary_ham10000.joblib")
        joblib.dump(scalers[model_name], f"/kaggle/working/{model_name}_scaler_rf_binary_ham10000.joblib")
        
        # Store results
        results[f'{model_name}_rf'] = {
            'model': rf_classifier,
            'scaler': scalers[model_name],
            'train_acc': train_accuracy,
            'val_acc': val_accuracy
        }
        
        # Release memory 
        release_memory()
    
    return results

def train_fusion_models(X_train_fused, y_train, X_val_fused, y_val):
    """Train models on fused features from all feature extractors"""
    log_message("\n" + "="*50)
    log_message("Training SVM classifier on fused features...")
    
    check_progress_update("Training SVM on fused features")
    
    # Scale fused features
    log_message("Scaling fused features...")
    scaler = StandardScaler()
    X_train_fused_scaled = scaler.fit_transform(X_train_fused)
    X_val_fused_scaled = scaler.transform(X_val_fused)
    
    # Save scaler
    joblib.dump(scaler, "/kaggle/working/fusion_scaler_svm_binary_ham10000.joblib")
    
    # Record start time
    start_time = time.time()
    
    #  SVM parameters 
    svm_classifier = SVC(
        kernel='rbf',
        probability=True,
        C=1.0,  
        gamma='scale',
        cache_size=3000,          
        max_iter=400,
        tol=1e-2,
        class_weight=None,  
        verbose=False,  
        decision_function_shape='ovo',
        break_ties=False
    )

     
    svm_classifier.fit(X_train_fused_scaled, y_train)
    
    # Calculate training time
    train_time = time.time() - start_time
    model_train_times['fusion_svm'] = train_time
    log_message(f"Fusion SVM training time: {train_time:.2f} seconds")
    
    # Training metrics
    y_train_pred = svm_classifier.predict(X_train_fused_scaled)
    train_accuracy = accuracy_score(y_train, y_train_pred)
    
    # Validation metrics
    y_val_pred = svm_classifier.predict(X_val_fused_scaled)
    val_accuracy = accuracy_score(y_val, y_val_pred)
    
    log_message(f"Fusion SVM Training Accuracy: {train_accuracy:.4f}")
    log_message(f"Fusion SVM Validation Accuracy: {val_accuracy:.4f}")
    
    # Save the model
    joblib.dump(svm_classifier, model_paths['svm'])
    log_message(f"Fusion SVM model saved to {model_paths['svm']}")
    
    # Train Random Forest
    log_message("\n" + "="*50)
    log_message("Training Random Forest classifier on fused features...")
    
    check_progress_update("Training Random Forest on fused features")
    
    # Record start time
    start_time = time.time()
    
    # Random Forest 
    rf_classifier = RandomForestClassifier(
        n_estimators=100,
        random_state=42,
        n_jobs=-1,
        max_depth=10,      
        min_samples_split=10,
        min_samples_leaf=4,
        bootstrap=True,    
        verbose=0,
        warm_start=False,
        max_features='sqrt',  
        criterion='gini',     
        max_leaf_nodes=100,
        class_weight=None  
    )
     
    rf_classifier.fit(X_train_fused_scaled, y_train)
    
    # Calculate training time
    train_time = time.time() - start_time
    model_train_times['fusion_rf'] = train_time
    log_message(f"Fusion Random Forest training time: {train_time:.2f} seconds")
    
    # Training metrics
    y_train_pred = rf_classifier.predict(X_train_fused_scaled)
    train_accuracy = accuracy_score(y_train, y_train_pred)
    
    # Validation metrics
    y_val_pred = rf_classifier.predict(X_val_fused_scaled)
    val_accuracy = accuracy_score(y_val, y_val_pred)
    
    log_message(f"Fusion Random Forest Training Accuracy: {train_accuracy:.4f}")
    log_message(f"Fusion Random Forest Validation Accuracy: {val_accuracy:.4f}")
    
    # Save the model and scaler
    joblib.dump(rf_classifier, model_paths['random_forest'])
    joblib.dump(scaler, "/kaggle/working/fusion_scaler_rf_binary_ham10000.joblib")
    log_message(f"Fusion Random Forest model saved to {model_paths['random_forest']}")
    
    return {
        'fusion_svm': {
            'model': svm_classifier,
            'scaler': scaler,
            'train_acc': train_accuracy,
            'val_acc': val_accuracy
        },
        'fusion_rf': {
            'model': rf_classifier,
            'scaler': scaler,
            'train_acc': train_accuracy,
            'val_acc': val_accuracy
        }
    }

def print_detailed_metrics(y_true, y_pred, y_prob, idx_to_class, model_name):
    """Print detailed metrics for a model including confusion matrix and classification report"""
    num_classes = len(idx_to_class)
    class_names = [idx_to_class[i] for i in range(num_classes)]
    
    # Generate classification report
    report = classification_report(y_true, y_pred, target_names=class_names, output_dict=True)
    
    # Extract precision, recall, and F1-score
    precision = {}
    recall = {}
    f1_score = {}
    support = {}
    
    for i, class_name in enumerate(class_names):
        precision[class_name] = report[class_name]['precision']
        recall[class_name] = report[class_name]['recall']
        f1_score[class_name] = report[class_name]['f1-score']
        support[class_name] = report[class_name]['support']
    
    # Create a DataFrame 
    metrics_df = pd.DataFrame({
        'Class': class_names,
        'Precision': [precision[cls] for cls in class_names],
        'Recall': [recall[cls] for cls in class_names],
        'F1-Score': [f1_score[cls] for cls in class_names],
        'Support': [support[cls] for cls in class_names]
    })
    
    # Print the detailed metrics
    log_message(f"\n{'-'*30} {model_name} Metrics {'-'*30}")
    log_message(f"Overall Accuracy: {report['accuracy']:.4f}")
    log_message(f"\nDetailed Metrics by Class:")
    log_message(metrics_df.to_string(index=False))
    
    # Print macro and weighted averages
    log_message(f"\nMacro Average:")
    log_message(f"  Precision: {report['macro avg']['precision']:.4f}")
    log_message(f"  Recall: {report['macro avg']['recall']:.4f}")
    log_message(f"  F1-Score: {report['macro avg']['f1-score']:.4f}")
    
    log_message(f"\nWeighted Average:")
    log_message(f"  Precision: {report['weighted avg']['precision']:.4f}")
    log_message(f"  Recall: {report['weighted avg']['recall']:.4f}")
    log_message(f"  F1-Score: {report['weighted avg']['f1-score']:.4f}")
    
    # Save metrics to global dictionary
    all_metrics[model_name] = {
        'accuracy': report['accuracy'],
        'precision': precision,
        'recall': recall,
        'f1_score': f1_score,
        'macro_avg_precision': report['macro avg']['precision'],
        'macro_avg_recall': report['macro avg']['recall'],
        'macro_avg_f1': report['macro avg']['f1-score'],
        'weighted_avg_precision': report['weighted avg']['precision'],
        'weighted_avg_recall': report['weighted avg']['recall'],
        'weighted_avg_f1': report['weighted avg']['f1-score']
    }
    
    # Plot confusion matrix
    cm = confusion_matrix(y_true, y_pred)
    
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=class_names, 
                yticklabels=class_names)
    plt.xlabel('Predicted Label', fontsize=14)
    plt.ylabel('True Label', fontsize=14)
    plt.title(f'Confusion Matrix - {model_name}', fontsize=16)
    
    plt.subplots_adjust(bottom=0.15, left=0.15, right=0.95, top=0.9)
    plt.savefig(f"/kaggle/working/{model_name}_confusion_matrix_binary_ham10000.png", dpi=300)
    plt.show()

def plot_roc_curves(y_test, y_test_prob, num_classes, idx_to_class, model_name):
    """Plot ROC curves for a model"""
    

    malignant_idx = [i for i, class_name in idx_to_class.items() if class_name == 'malignant'][0]
    
    # Get probabilities for the malignant class
    malignant_prob = y_test_prob[:, malignant_idx]
    
    # Create binary labels where 1 = malignant, 0 = benign
    y_test_binary = (y_test == malignant_idx).astype(int)
    
    # Compute ROC curve and ROC area
    fpr, tpr, _ = roc_curve(y_test_binary, malignant_prob)
    roc_auc = auc(fpr, tpr)
    
    # Plot ROC curve
    plt.figure(figsize=(10, 8))
    
    plt.plot(fpr, tpr, 
             label=f'ROC curve (area = {roc_auc:.4f})',
             color='darkorange', linewidth=2)
    
    # Add diagonal line
    plt.plot([0, 1], [0, 1], 'k--', lw=2)
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate', fontsize=14)
    plt.ylabel('True Positive Rate', fontsize=14)
    plt.title(f'ROC Curve for {model_name} (Binary Classification)', fontsize=16)
    plt.legend(loc="lower right", fontsize=12)
    plt.grid(alpha=0.3)
    
    plt.subplots_adjust(bottom=0.1, left=0.1, right=0.9, top=0.9)
    plt.savefig(f"/kaggle/working/{model_name}_roc_curve_binary_ham10000.png", dpi=300)
    plt.show()
    
  

def evaluate_individual_models(models_dict, X_test_dict, y_test, idx_to_class, num_classes):
    """Evaluate individual models and print detailed metrics"""
    results = {}
    
    for model_key, model_info in models_dict.items():
        if model_key.startswith('fusion'):
            continue  # Skip fusion models for now
            
        # Extract model name and type
        model_parts = model_key.split('_')
        feature_type = model_parts[0]
        classifier_type = model_parts[1]
        
        log_message(f"\nEvaluating {classifier_type.upper()} with {feature_type} features...")
        
        check_progress_update(f"Evaluating {classifier_type.upper()} with {feature_type} features")
        
        # Record start time
        start_time = time.time()
        
        # Get the model, scaler and features
        model = model_info['model']
        scaler = model_info['scaler']
        X_test_scaled = scaler.transform(X_test_dict[feature_type])
        
        # Make predictions
        y_test_pred = model.predict(X_test_scaled)
        test_accuracy = accuracy_score(y_test, y_test_pred)
        
        # Get probabilities
        y_test_prob = model.predict_proba(X_test_scaled)
        
        # Calculate evaluation time
        eval_time = time.time() - start_time
        model_eval_times[model_key] = eval_time
        model_total_times[model_key] = model_train_times[model_key] + eval_time
        
        log_message(f"{feature_type} {classifier_type.upper()} Test Accuracy: {test_accuracy:.4f}")
        log_message(f"Evaluation Time: {eval_time:.2f} seconds")
        
        # Print metrics and plot confusion matrix
        print_detailed_metrics(y_test, y_test_pred, y_test_prob, idx_to_class, model_key)
        
        # Plot ROC curves
        plot_roc_curves(y_test, y_test_prob, num_classes, idx_to_class, model_key)
        
        # Store results
        results[model_key] = {
            'test_acc': test_accuracy,
            'eval_time': eval_time,
            'predictions': y_test_pred,
            'probabilities': y_test_prob
        }
    
    return results

def evaluate_fusion_models(models_dict, X_test_fused, y_test, idx_to_class, num_classes):
    """Evaluate fusion models and print detailed metrics"""
    results = {}
    
    for model_key in ['fusion_svm', 'fusion_rf']:
        model_info = models_dict[model_key]
        classifier_type = model_key.split('_')[1].upper()
        
        log_message(f"\nEvaluating Fusion {classifier_type}...")
        
        check_progress_update(f"Evaluating Fusion {classifier_type}")
        
        # Record start time
        start_time = time.time()
        
        # Get the model and scaler
        model = model_info['model']
        scaler = model_info['scaler']
        X_test_fused_scaled = scaler.transform(X_test_fused)
        
        # Make predictions
        y_test_pred = model.predict(X_test_fused_scaled)
        test_accuracy = accuracy_score(y_test, y_test_pred)
        
        # Get probabilities
        y_test_prob = model.predict_proba(X_test_fused_scaled)
        
        # Calculate evaluation time
        eval_time = time.time() - start_time
        model_eval_times[model_key] = eval_time
        model_total_times[model_key] = model_train_times[model_key] + eval_time
        
        log_message(f"Fusion {classifier_type} Test Accuracy: {test_accuracy:.4f}")
        log_message(f"Evaluation Time: {eval_time:.2f} seconds")
        
        # Print metrics and plot confusion matrix
        print_detailed_metrics(y_test, y_test_pred, y_test_prob, idx_to_class, model_key)
        
        # Plot ROC curves
        plot_roc_curves(y_test, y_test_prob, num_classes, idx_to_class, model_key)
        
        # Store results
        results[model_key] = {
            'test_acc': test_accuracy,
            'eval_time': eval_time,
            'predictions': y_test_pred,
            'probabilities': y_test_prob
        }
    
    return results

def compare_all_models(all_results):
    """Compare performance of all models"""
    # Collect test accuracies
    model_names = []
    test_accuracies = []
    
    for model_key, results in all_results.items():
        model_names.append(model_key)
        test_accuracies.append(results['test_acc'])
    
    # Create dataframe 
    comparison_df = pd.DataFrame({
        'Model': model_names,
        'Test Accuracy': test_accuracies
    })
    
    # Add more metrics from the all_metrics dictionary
    comparison_df['Macro Avg F1'] = [all_metrics[model]['macro_avg_f1'] for model in model_names]
    comparison_df['Training Time (s)'] = [model_train_times.get(model, 0) for model in model_names]
    comparison_df['Evaluation Time (s)'] = [model_eval_times.get(model, 0) for model in model_names]
    comparison_df['Total Time (s)'] = [model_total_times.get(model, 0) for model in model_names]
    
    # Sort by accuracy
    comparison_df = comparison_df.sort_values(by='Test Accuracy', ascending=False)
    
    log_message("\nModel Performance Comparison:")
    log_message(comparison_df.to_string(index=False))
    
    # Save comparison 
    comparison_df.to_csv('/kaggle/working/model_comparison_binary_ham10000.csv', index=False)
    
    # Plot comparison chart for accuracy
    plt.figure(figsize=(14, 8))
    
  
    colors = ['lightgreen' if 'fusion' in name else 'skyblue' for name in comparison_df['Model']]
    
    # Create the bars with labels
    bars = plt.bar(comparison_df['Model'], comparison_df['Test Accuracy'], color=colors)
    
    # Add values on top of bars
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height + 0.005,
                f'{height:.4f}', ha='center', va='bottom', fontweight='bold')
    
    plt.xlabel('Models', fontsize=14)
    plt.ylabel('Test Accuracy', fontsize=14)
    plt.title('Test Accuracy Comparison: Binary Classification (Malignant vs Benign)', fontsize=16)
    plt.xticks(rotation=45, ha='right')
    plt.ylim(0, max(test_accuracies) + 0.05)
    plt.grid(axis='y', linestyle='--', alpha=0.3)
    
    # Create proper legend
    fusion_patch = plt.Rectangle((0,0),1,1, color='lightgreen', label='Fusion Models')
    individual_patch = plt.Rectangle((0,0),1,1, color='skyblue', label='Individual Models')
    plt.legend(handles=[fusion_patch, individual_patch])
    
    plt.subplots_adjust(bottom=0.25, left=0.1, right=0.95, top=0.9)
    plt.savefig('/kaggle/working/all_models_accuracy_comparison_binary_ham10000.png', dpi=300)
    plt.show()
    
    return comparison_df

# Define the main function 
def main():
    # Initialize progress update time
    global last_progress_update
    last_progress_update = time.time()
    
    # Set device for feature extraction
    device, device_type = setup_device()
    log_message(f"Using device: {device}")
    
    # Apply optimizations to PyTorch
    if device_type == 'cuda':
        torch.backends.cudnn.benchmark = True
        torch.backends.cuda.matmul.allow_tf32 = True
        torch.backends.cudnn.allow_tf32 = True
    elif device_type == 'tpu':
        log_message("Using TPU-specific optimizations")
    
    # Dataset path
    dataset_path = "/kaggle/input/skin-cancer-mnist-ham10000/"
    
    # Run augmentation and preparation 
    log_message("Starting dataset preparation...")
    dataset_info = augment_and_prepare_dataset(dataset_path)
    
    # Extract features for training set
    log_message("\nExtracting features from training set...")
    X_train_dict, X_train_fused, y_train = extract_features_and_labels(dataset_info['train_loader'], device, device_type)
    
    # Extract features for validation set
    log_message("\nExtracting features from validation set...")
    X_val_dict, X_val_fused, y_val = extract_features_and_labels(dataset_info['val_loader'], device, device_type)
    
    # Extract features for test set
    log_message("\nExtracting features from test set...")
    X_test_dict, X_test_fused, y_test = extract_features_and_labels(dataset_info['test_loader'], device, device_type)
    
    # Get number of classes
    num_classes = dataset_info['num_classes']
    log_message(f"\nNumber of classes: {num_classes}")
    assert num_classes == 2, "Expected 2 classes for binary classification"
    
    # Train individual models for each feature extractor
    log_message("\nTraining individual models...")
    individual_models = train_individual_models(X_train_dict, y_train, y_val, X_val_dict)
    
    # Train models on fused features
    log_message("\nTraining fusion models...")
    fusion_models = train_fusion_models(X_train_fused, y_train, X_val_fused, y_val)
    
    # Combine all models for evaluation
    all_models = {**individual_models, **fusion_models}
    
    # Evaluate individual models
    log_message("\nEvaluating individual models...")
    individual_results = evaluate_individual_models(all_models, X_test_dict, y_test, dataset_info['idx_to_class'], num_classes)
    
    # Evaluate fusion models
    log_message("\nEvaluating fusion models...")
    fusion_results = evaluate_fusion_models(all_models, X_test_fused, y_test, dataset_info['idx_to_class'], num_classes)
    
    # Combine all results
    all_results = {**individual_results, **fusion_results}
    
    # Compare all models
    log_message("\nComparing all models...")
    comparison = compare_all_models(all_results)
    
    
  
  
    
   
  
    
    
    
    return comparison

if __name__ == "__main__":
    # Fix for SSL certificate issue
    ssl._create_default_https_context = ssl._create_unverified_context

    # For 10-minute progress updates
    last_progress_update = time.time()

    def check_progress_update(message):
        """Check if 10 minutes have passed to update progress"""
        global last_progress_update
        current_time = time.time()
        
        if current_time - last_progress_update >= 600:  # 10 minutes = 600 seconds
            log_message(f"Progress update: {message}")
            last_progress_update = current_time
            return True
        return False

    # Determine optimal thread count based on system
    cpu_count = multiprocessing.cpu_count()
    optimal_threads = max(1, cpu_count - 1)  

    # Set environment variables for optimized performance
    os.environ['OMP_NUM_THREADS'] = str(optimal_threads)
    os.environ['MKL_NUM_THREADS'] = str(optimal_threads)
    os.environ['NUMEXPR_NUM_THREADS'] = str(optimal_threads)
    os.environ['VECLIB_MAXIMUM_THREADS'] = str(optimal_threads)
    os.environ['OPENBLAS_NUM_THREADS'] = str(optimal_threads)
    
    # Dictionary to store model paths
    model_paths = {
        'svm': "/kaggle/working/feature_fusion_svm_binary_ham10000.joblib",
        'random_forest': "/kaggle/working/feature_fusion_rf_binary_ham10000.joblib"
    }

    # Dictionaries to store metrics
    model_train_times = {}
    model_eval_times = {}
    model_total_times = {}
    all_metrics = {}
    
    # Set random seeds for reproducibility
    torch.manual_seed(42)
    random.seed(42)
    np.random.seed(42)
    
    try:
        main()
    except Exception as e:
        log_message(f"ERROR: {str(e)}")
        # Save stack trace to log file
        import traceback
        log_message(traceback.format_exc())
```

```python
import os
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, models
from PIL import Image
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    accuracy_score, classification_report, confusion_matrix, 
    roc_curve, auc, precision_recall_curve, average_precision_score
)
from sklearn.preprocessing import label_binarize, StandardScaler
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from tqdm import tqdm
import time
import random
import joblib
import ssl
import gc  
import multiprocessing
from concurrent.futures import ThreadPoolExecutor, ProcessPoolExecutor
import datetime

# Fix for SSL certificate issue
ssl._create_default_https_context = ssl._create_unverified_context

# Function to log messages to both console and file
def log_message(message, log_file="/kaggle/working/progress_log.txt"):
    timestamp = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    formatted_message = f"[{timestamp}] {message}"
    
    # Print to console
    print(formatted_message)
    
    # Write to log file
    with open(log_file, "a") as f:
        f.write(formatted_message + "\n")

# Determine optimal thread count based on system
cpu_count = multiprocessing.cpu_count()
optimal_threads = max(1, cpu_count - 1)  

# Set environment variables for optimized performance
os.environ['OMP_NUM_THREADS'] = str(optimal_threads)
os.environ['MKL_NUM_THREADS'] = str(optimal_threads)
os.environ['NUMEXPR_NUM_THREADS'] = str(optimal_threads)
os.environ['VECLIB_MAXIMUM_THREADS'] = str(optimal_threads)
os.environ['OPENBLAS_NUM_THREADS'] = str(optimal_threads) 

# Function to setup device (TPU/GPU/CPU)
def setup_device():
    try:
        import torch_xla
        import torch_xla.core.xla_model as xm
        device = xm.xla_device()
        log_message("Using TPU for computation")
        return device, "tpu"
    except ImportError:
        if torch.cuda.is_available():
            torch.cuda.set_device(0)
            log_message(f"Using GPU: {torch.cuda.get_device_name(0)}")
            return torch.device("cuda"), "cuda"
        else:
            log_message("Using CPU for computation")
            return torch.device("cpu"), "cpu"

# Set up device
device, device_type = setup_device()

# Optimize CUDA operations if using GPU
if device_type == 'cuda':
    # Optimize CUDA operations
    torch.backends.cudnn.benchmark = True
    torch.backends.cudnn.deterministic = False
    torch.backends.cuda.matmul.allow_tf32 = True
    torch.backends.cudnn.allow_tf32 = True
    
    # Clear CUDA cache
    torch.cuda.empty_cache()
    
    # Monitor memory
    torch.cuda.memory.memory_stats() 
    torch.cuda.memory.reset_peak_memory_stats()

# Use mixed precision training 
if device_type == 'cuda' and torch.cuda.get_device_capability()[0] >= 7:
    log_message("Enabling mixed precision training")
    torch.backends.cuda.matmul.allow_fp16_reduced_precision_reduction = True
    
    from torch.cuda.amp import autocast, GradScaler
    use_amp = True
else:
    use_amp = False

# Memory management
def release_memory(verbose=False):
    """Release unused memory in PyTorch and trigger garbage collection"""
    before = 0
    if torch.cuda.is_available() and verbose:
        before = torch.cuda.memory_allocated()
        
    if torch.cuda.is_available():
        torch.cuda.empty_cache()
        
    collected = gc.collect(generation=2)  
    
    if torch.cuda.is_available() and verbose:
        after = torch.cuda.memory_allocated()
        log_message(f"Memory freed: {(before-after)/1024**2:.2f} MB, Objects collected: {collected}")

# Initialize progress update timer
last_progress_update = time.time()

def check_progress_update(message):
    """Check if 10 minutes have passed to update progress"""
    global last_progress_update
    current_time = time.time()
    
    if current_time - last_progress_update >= 600:  # 10 minutes = 600 seconds
        log_message(f"Progress update: {message}")
        last_progress_update = current_time
        return True
    return False

plt.ioff()  
plt.rcParams['figure.dpi'] = 100  
plt.style.use('fast') 

# Optimize NumPy operations
np.seterr(divide='ignore', invalid='ignore') 
np.set_printoptions(precision=4, suppress=True)  

# Set random seeds 
torch.manual_seed(42)
random.seed(42)
np.random.seed(42)

# Define target counts for each class 
target_counts = {
    'akiec': 6705,
    'bcc': 6705,
    'bkl': 6705,
    'df': 6705,
    'mel': 6705,
    'nv':6705,
    'vasc': 6705
}

# Dictionary to store model paths
model_paths = {
    'svm': "/kaggle/working/feature_fusion_svm_multi_ham10000.joblib",
    'random_forest': "/kaggle/working/feature_fusion_rf_multi_ham10000.joblib"
}

# Dictionaries to store metrics
model_train_times = {}
model_eval_times = {}
model_total_times = {}
all_metrics = {}

# DATA PROCESSING FUNCTIONS 

def augment_and_prepare_dataset(dataset_path):
    log_message("\nLoading and augmenting dataset...")
    
    # Load metadata 
    metadata_file = os.path.join(dataset_path, "HAM10000_metadata.csv")
    metadata = pd.read_csv(metadata_file, low_memory=False)
    
    # Calculate initial class distribution
    initial_class_counts = metadata['dx'].value_counts()
    log_message("\nInitial class distribution:")
    for cls, count in initial_class_counts.items():
        log_message(f"{cls}: {count}")
    
    # Get image paths -
    image_folders = ["HAM10000_images_part_1", "HAM10000_images_part_2"]
    image_path_dict = {}
    
    check_progress_update("Processing image folders...")
    
    def process_folder(folder):
        local_dict = {}
        folder_path = os.path.join(dataset_path, folder)
        if os.path.exists(folder_path):
            for img_name in os.listdir(folder_path):
                if img_name.endswith('.jpg'):
                    image_id = os.path.splitext(img_name)[0]
                    local_dict[image_id] = os.path.join(folder_path, img_name)
        return local_dict
    
    # Process folders in parallel 
    if len(image_folders) > 1:
        with ThreadPoolExecutor(max_workers=len(image_folders)) as executor:
            results = list(executor.map(process_folder, image_folders))
            
        for result in results:
            image_path_dict.update(result)
    else:
        image_path_dict = process_folder(image_folders[0])
    
    # Add image paths to metadata 
    metadata['image_path'] = metadata['image_id'].map(image_path_dict)
    
    # Create a plot to visualize initial class distribution
    plt.figure(figsize=(12, 6))
    ax = sns.barplot(x=initial_class_counts.index, y=initial_class_counts.values)
    plt.title('Initial Class Distribution', fontsize=15)
    plt.xlabel('Skin Lesion Class', fontsize=12)
    plt.ylabel('Number of Images', fontsize=12)
    
    # Add count labels above the bars
    for i, count in enumerate(initial_class_counts.values):
        ax.text(i, count + 50, str(count), ha='center', fontsize=10, fontweight='bold')
    
    plt.subplots_adjust(bottom=0.15, top=0.9)
    plt.savefig("/kaggle/working/initial_class_distribution.png", dpi=300)
    plt.close()  
    
    check_progress_update("Downsampling classes...")
    
    # Downsampling classes 
    processed_metadata = []
    
    # For each class, downsample if needed 
    for class_name in metadata['dx'].unique():
        class_samples = metadata[metadata['dx'] == class_name]
        target_count = target_counts[class_name]
        
        if len(class_samples) > target_count:
            # Downsample class to target count
            log_message(f"Downsampling '{class_name}' class from {len(class_samples)} to {target_count} samples")
            class_samples = class_samples.sample(target_count, random_state=42)
        
        processed_metadata.append(class_samples)
    
    # Combine all processed classes
    processed_metadata = pd.concat(processed_metadata, ignore_index=True)
    
    # Calculate new class distribution after downsampling
    processed_class_counts = processed_metadata['dx'].value_counts()
    log_message("\nClass distribution after downsampling:")
    for cls, count in processed_class_counts.items():
        log_message(f"{cls}: {count}")
    
    # Define augmentation transforms 
    augmentation_transforms = [
        transforms.RandomRotation(15),  # 15 degree rotation
        transforms.RandomHorizontalFlip(p=0.5),  # Horizontal flip
        transforms.RandomVerticalFlip(p=0.5),  # Vertical flip
        transforms.RandomAffine(
            degrees=0,  # No additional rotation
            translate=(0.1, 0.1),  # Width and height shift of 10%
            scale=(0.8, 1.2),  # Zoom effect with scale range of 0.2 (20%)
        ),
    ]
    
    # Use the same augmentations but at different probabilities for diversity
    strong_augmentation_transforms = [
        transforms.RandomRotation(15),  # 15 degree rotation
        transforms.RandomHorizontalFlip(p=0.8),  # Horizontal flip 
        transforms.RandomVerticalFlip(p=0.8),  # Vertical flip 
        transforms.RandomAffine(
            degrees=0,  # No additional rotation
            translate=(0.1, 0.1),  # Width and height shift of 10%
            scale=(0.8, 1.2),  # Zoom effect with scale range of 0.2 (20%)
        ),
    ]
    
    # STEP 1: Create all images (original + augmented) as PIL Images
    log_message("\nSTEP 1: Processing original images and generating augmentations...")
    all_images = []  
    
    batch_size = 128  # Process images in batches
    
    # Create a function to process a batch of images
    def process_image_batch(batch_rows):
        batch_results = []
        for _, row in batch_rows.iterrows():
            try:
                img_path = row['image_path']
                label = row['dx']  
                image_id = row['image_id']
                
                if pd.isna(img_path):
                    continue
                    
                # Load original image and add to the list
                img = Image.open(img_path).convert('RGB')
                batch_results.append((img, label, f"{image_id}_orig"))
                
            except Exception as e:
                log_message(f"Error processing image {row['image_id']}: {e}")
        return batch_results
    
    # Process original images in batches
    num_batches = len(processed_metadata) // batch_size + (1 if len(processed_metadata) % batch_size > 0 else 0)
    
    for i in tqdm(range(num_batches), desc="Processing original images"):
        check_progress_update(f"Processing original images - batch {i+1}/{num_batches}")
        
        start_idx = i * batch_size
        end_idx = min((i + 1) * batch_size, len(processed_metadata))
        batch = processed_metadata.iloc[start_idx:end_idx]
        all_images.extend(process_image_batch(batch))
        
        # Release memory 
        if i % 10 == 0:
            release_memory()
    
    # Create augmented data
    log_message("\nGenerating augmented images...")
    
    for class_name, target_count in target_counts.items():
        # Get original samples for this class
        class_samples = processed_metadata[processed_metadata['dx'] == class_name]
        current_count = len(class_samples)
        
        check_progress_update(f"Processing class {class_name} - current count: {current_count}, target: {target_count}")
        
        if current_count >= target_count:
            log_message(f"Class '{class_name}' already has {current_count} samples, no augmentation needed")
            continue
            
        # Calculate how many augmented samples we need
        num_augmented_needed = target_count - current_count
        
        log_message(f"Augmenting '{class_name}' class: adding {num_augmented_needed} samples to reach {target_count}")
        
        # Generate augmented samples 
        num_per_original = max(1, num_augmented_needed // len(class_samples) + 1)
        counter = 0
        
        # Pre-load images 
        preloaded_images = {}
        for _, row in tqdm(class_samples.iterrows(), desc=f"Preloading images for {class_name}", total=len(class_samples)):
            try:
                img_path = row['image_path']
                if pd.isna(img_path):
                    continue
                preloaded_images[row['image_id']] = Image.open(img_path).convert('RGB')
            except Exception as e:
                log_message(f"Error preloading image {row['image_id']}: {e}")
        
        augmentation_count = 0
        indices = list(range(len(class_samples)))
        
        while counter < num_augmented_needed:
            # Check for progress update
            check_progress_update(f"Augmenting class {class_name}: {counter}/{num_augmented_needed} samples generated")
            
            # Cycle through the indices
            idx = indices[augmentation_count % len(indices)]
            sample = class_samples.iloc[idx]
            image_id = sample['image_id']
            
            try:
                # Get the preloaded image
                if image_id not in preloaded_images:
                    augmentation_count += 1
                    continue
                    
                img = preloaded_images[image_id]
                
                # Decide which augmentation strategy to use
                if counter % 5 == 0:  # Every 5th augmentation, use strong mix
                    # Apply the strong augmentations
                    aug_img = img
                    for transform in strong_augmentation_transforms:
                        aug_img = transform(aug_img)
                
                else:
                    # Apply the standard augmentations
                    aug_img = img
                    for transform in augmentation_transforms:
                        aug_img = transform(aug_img)
                
                # Add augmented image to the list
                all_images.append((aug_img, class_name, f"{sample['image_id']}_aug_{counter}"))
                counter += 1
                
                if counter % 3000 == 0:
                    log_message(f"Generated {counter}/{num_augmented_needed} augmented samples for class {class_name}")
                    release_memory()
                    
            except Exception as e:
                log_message(f"Error generating augmented image: {e}")
            
            augmentation_count += 1
    
    # Clean up preloaded images to free memory
    preloaded_images = None
    release_memory()
    
    # STEP 2: Create transforms 
    log_message(f"\nSTEP 2: Creating transforms ...")
    
    check_progress_update("Creating image transforms...")
    
    # Define transform for all models
    transforms_dict = {
        'mobilenet': transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ]),
        'densenet121': transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ]),
        'resnet50': transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
    }
    # Process all images with the common transform for dataset splitting
    all_processed_data = []
    # Use MobileNetV3 transforms as default for splitting purposes
    default_transform = transforms_dict['mobilenet']
    
    log_message(f"Applying resize and normalize for dataset splitting...")
    for i, (img, label, image_id) in enumerate(tqdm(all_images, desc="Processing images")):
        # Check for progress updates
        if i % 100 == 0:
            check_progress_update(f"Processing images: {i}/{len(all_images)}")
            
        img_tensor = default_transform(img)
        all_processed_data.append((img_tensor, label, image_id, img))
        
    # STEP 3: Split the dataset
    log_message("\nSTEP 3: Splitting dataset into train, validation, and test sets...")
    
    check_progress_update("Splitting dataset...")
    
    # Create class to index mapping for multi-class classification
    classes = sorted(list(set([label for _, label, _ in all_images])))
    class_to_idx = {cls: i for i, cls in enumerate(classes)}
    log_message(f"Multi-class mapping: {class_to_idx}")
    
    # Convert string labels to indices
    processed_data = [(img, class_to_idx[label], image_id) for img, label, image_id in all_images]
    
    # Determine the exact split sizes 
    total_size = len(processed_data)
    train_size = int(0.70 * total_size)
    remaining = total_size - train_size
    val_size = remaining // 2  
    test_size = remaining - val_size  
    
    if val_size != test_size:
        train_size += 1
        val_size = test_size  # Now they're exactly equal
        
    log_message(f"Splitting dataset: {train_size} train, {val_size} validation, {test_size} test samples")
    
    # Stratified splitting to maintain class distribution
    random.seed(42)
    random.shuffle(processed_data)
    
    # Extract labels 
    labels = [label for _, label, _ in processed_data]
    
    indices = np.arange(total_size)
    train_indices, temp_indices = train_test_split(
        indices, test_size=val_size+test_size, 
        stratify=labels, 
        random_state=42
    )
    
    val_indices, test_indices = train_test_split(
        temp_indices, 
        test_size=0.5,  # Exactly half for each
        stratify=[labels[i] for i in temp_indices],
        random_state=42
    )
    
    # Create datasets
    train_data = [processed_data[i] for i in train_indices]
    val_data = [processed_data[i] for i in val_indices]
    test_data = [processed_data[i] for i in test_indices]
    
    check_progress_update("Creating datasets...")
    
    # Create  dataset class for feature extraction
    class SkinLesionDataset(Dataset):
        def __init__(self, data, transform=None):
            self.data = data
            self.transform = transform
            
        def __len__(self):
            return len(self.data)
            
        def __getitem__(self, idx):
            img, label, _ = self.data[idx]
            if self.transform:
                img = self.transform(img)
            return img, label
    
    # Optimize batch_size and num_workers based on system resources
    if device_type == 'tpu':
        batch_size = 128
        num_workers = 96
    elif device_type == 'cuda':
        gpu_memory = torch.cuda.get_device_properties(0).total_memory / 1e9
        
        if gpu_memory > 16:  
            batch_size = 512
        elif gpu_memory > 8:  
            batch_size = 256
        else:  
            batch_size = 128
        
        num_workers = min(128, os.cpu_count() or 4)
    else:
        # Default batch size for CPU
        batch_size = 128
        num_workers = min(96, os.cpu_count() or 4)
    
    log_message(f"Using batch_size={batch_size}, num_workers={num_workers}")
    
    # Create optimized data loaders 
    train_dataset = SkinLesionDataset(train_data, transform=default_transform)
    val_dataset = SkinLesionDataset(val_data, transform=default_transform)
    test_dataset = SkinLesionDataset(test_data, transform=default_transform)
    
    train_loader = DataLoader(
        train_dataset, 
        batch_size=batch_size, 
        shuffle=False, 
        num_workers=num_workers, 
        pin_memory=(device_type != 'tpu'),
        persistent_workers=(num_workers > 0 and device_type != 'tpu'),
        prefetch_factor=2 if num_workers > 0 else None
    )
    
    val_loader = DataLoader(
        val_dataset, 
        batch_size=batch_size, 
        shuffle=False, 
        num_workers=num_workers, 
        pin_memory=(device_type != 'tpu'),
        persistent_workers=(num_workers > 0 and device_type != 'tpu'),
        prefetch_factor=2 if num_workers > 0 else None
    )
    
    test_loader = DataLoader(
        test_dataset, 
        batch_size=batch_size, 
        shuffle=False, 
        num_workers=num_workers, 
        pin_memory=(device_type != 'tpu'),
        persistent_workers=(num_workers > 0 and device_type != 'tpu'),
        prefetch_factor=2 if num_workers > 0 else None
    )
    
    # Count samples in each split by class
    def count_classes(data_list):
        count_dict = {}
        for _, label, _ in data_list:
            if label not in count_dict:
                count_dict[label] = 0
            count_dict[label] += 1
        return count_dict
    
    train_counts = count_classes(train_data)
    val_counts = count_classes(val_data)
    test_counts = count_classes(test_data)
    
    # Convert class indices back to names for display
    idx_to_class = {v: k for k, v in class_to_idx.items()}
    
    # Display class distribution in each split
    log_message("\nClass distribution in train split:")
    for label, count in train_counts.items():
        log_message(f"  Class {idx_to_class[label]}: {count} images")
    
    log_message("\nClass distribution in validation split:")
    for label, count in val_counts.items():
        log_message(f"  Class {idx_to_class[label]}: {count} images")
    
    log_message("\nClass distribution in test split:")
    for label, count in test_counts.items():
        log_message(f"  Class {idx_to_class[label]}: {count} images")
    
    # Clear processed data to free memory
    processed_data = None
    all_images = None
    release_memory()
    
    return {
        'train_loader': train_loader,
        'val_loader': val_loader,
        'test_loader': test_loader,
        'class_to_idx': class_to_idx,
        'idx_to_class': idx_to_class,
        'classes': classes,
        'num_classes': len(classes)
    }

class MobileNetV3Extractor(nn.Module):
    """Feature extractor using MobileNetV3 without the classifier"""
    def __init__(self):
        super(MobileNetV3Extractor, self).__init__()
      
        model = models.mobilenet_v3_large(weights=models.MobileNet_V3_Large_Weights.IMAGENET1K_V1)
        self.features = nn.Sequential(*list(model.children())[:-1])
        
    def forward(self, x):
        x = self.features(x)
        return x.squeeze()

class DenseNet121Extractor(nn.Module):
    """Feature extractor using DenseNet121 without the classifier"""
    def __init__(self):
        super(DenseNet121Extractor, self).__init__()
       
        model = models.densenet121(weights=models.DenseNet121_Weights.IMAGENET1K_V1)
        self.features = nn.Sequential(*list(model.children())[:-1])
        
    def forward(self, x):
        x = self.features(x)
        # Apply global average pooling
        x = torch.nn.functional.adaptive_avg_pool2d(x, (1, 1))
        return x.squeeze()

class ResNet50Extractor(nn.Module):
    """Feature extractor using ResNet50 without the classifier"""
    def __init__(self):
        super(ResNet50Extractor, self).__init__()
        
        model = models.resnet50(weights=models.ResNet50_Weights.IMAGENET1K_V1)
        self.features = nn.Sequential(*list(model.children())[:-1])
        
    def forward(self, x):
        x = self.features(x)
        return x.squeeze()

# FEATURE EXTRACTION FUNCTIONS 

def extract_features_from_model(model, data_loader, device, device_type, model_name):
    """Extract features using a specific model with optimized batch processing"""
    model.to(device)
    model.eval()
    
    features = []
    batch_count = 0
    
    with torch.no_grad():
        for i, (images, _) in enumerate(tqdm(data_loader, desc=f"Extracting {model_name} features")):
            # Check progress update
            if i % 5 == 0:
                check_progress_update(f"Extracting {model_name} features: batch {i}/{len(data_loader)}")
                
            # Handle TPU differently
            if device_type == 'tpu':
                import torch_xla.core.xla_model as xm
                images = images.to(device)
                batch_features = model(images)
                # Move features to CPU
                batch_features = xm.all_gather(batch_features).cpu().numpy()
            else:
                images = images.to(device, non_blocking=True)
                
                if use_amp and device_type == 'cuda':
                    with autocast():
                        batch_features = model(images)
                else:
                    batch_features = model(images)
                    
                batch_features = batch_features.cpu().numpy()
                
            features.append(batch_features)
            
            batch_count += 1
            if batch_count % 20 == 0: 
                if device_type == 'cuda':
                    torch.cuda.empty_cache()
    
    # stack all features
    return np.vstack(features)

# extract features
def extract_features_and_labels(data_loader, device, device_type):
    """Extract features from all models separately and combine with optimized memory handling"""
   
    log_message("\nInitializing feature extractors...")
    mobilenet_extractor = MobileNetV3Extractor()
    
    # Extract labels  
    labels = []
    for _, targets in data_loader:
        if device_type == 'tpu':
            labels.append(targets.cpu().numpy())
        else:
            labels.append(targets.numpy())
    labels = np.concatenate(labels)
    
    # Extract features from each model separately
    log_message("\nExtracting features from MobileNetV3...")
    mobilenet_features = extract_features_from_model(mobilenet_extractor, data_loader, device, device_type, "MobileNetV3")
    log_message(f"MobileNetV3 features shape: {mobilenet_features.shape}")
    
    # Clear the first model to free up GPU memory
    mobilenet_extractor = None
    release_memory()
    
    check_progress_update("Starting DenseNet121 feature extraction")
    
    log_message("\nExtracting features from DenseNet121...")
    densenet_extractor = DenseNet121Extractor()
    densenet_features = extract_features_from_model(densenet_extractor, data_loader, device, device_type, "DenseNet121")
    log_message(f"DenseNet121 features shape: {densenet_features.shape}")
    
    # Clear the second model to free up GPU memory
    densenet_extractor = None
    release_memory()
    
    check_progress_update("Starting ResNet50 feature extraction")
    
    log_message("\nExtracting features from ResNet50...")
    resnet50_extractor = ResNet50Extractor()
    resnet50_features = extract_features_from_model(resnet50_extractor, data_loader, device, device_type, "ResNet50")
    log_message(f"ResNet50 features shape: {resnet50_features.shape}")
    
    # Clear the third model to free up GPU memory
    resnet50_extractor = None
    release_memory()
    
    feature_dict = {
        'mobilenet': mobilenet_features,
        'densenet': densenet_features,
        'resnet50': resnet50_features
    }
    
    # create fused features by concatenating
    fused_features = np.concatenate([
        mobilenet_features, 
        densenet_features, 
        resnet50_features
    ], axis=1)
    
    log_message(f"\nFused features shape: {fused_features.shape}")
    
    # Print label distribution
    unique_labels, counts = np.unique(labels, return_counts=True)
    log_message("\nLabel distribution in this split:")
    for label, count in zip(unique_labels, counts):
        log_message(f"  Class {label}: {count} images")
    
    # Convert to float32 to reduce memory usage
    for key in feature_dict:
        feature_dict[key] = feature_dict[key].astype(np.float32)
    
    fused_features = fused_features.astype(np.float32)
    
    return feature_dict, fused_features, labels

# MODEL TRAINING FUNCTIONS 

def train_individual_models(feature_dict, y_train, y_val, X_val_dict):
    """Train individual models with features from each feature extractor with enhanced parameters"""
    results = {}
    
    # Scale the features 
    scalers = {}
    scaled_features = {}
    scaled_val_features = {}
    
    # Scale each feature set separately
    for model_name, features in feature_dict.items():
        log_message(f"Scaling features for {model_name}...")
        scaler = StandardScaler()
        scaled_features[model_name] = scaler.fit_transform(features)
        scaled_val_features[model_name] = scaler.transform(X_val_dict[model_name])
        scalers[model_name] = scaler
    
    #  SVM parameters
    SVM_PARAMS = {
        'kernel': 'rbf',
        'probability': True,
        'C': 1.0,
        'gamma': 'scale',  
        'cache_size': 3000,  
        'max_iter': 400,   
        'tol': 1e-2,        
        'class_weight': None,
        'verbose': False,    
        'decision_function_shape': 'ovr',  
        'break_ties': False  
     }
     
    #  RF parameters 
    RF_PARAMS = {
        'n_estimators': 100,     
        'random_state': 42,
        'n_jobs': -1,            
        'max_depth': 10,        
        'min_samples_split': 10, 
        'min_samples_leaf': 4,   
        'bootstrap': True,      
        'verbose': 0,           
        'warm_start': False,
        'max_features': 'sqrt',  
        'criterion': 'gini',     
        'max_leaf_nodes': 100    
    }
    
    for model_name, features in scaled_features.items():
        log_message(f"\n{'='*50}")
        log_message(f"Training SVM using {model_name} features...")
        
        check_progress_update(f"Training SVM with {model_name} features")
        
        # Record start time
        start_time = time.time()
        
        # Train SVM with optimized parameters
        svm_classifier = SVC(**SVM_PARAMS)
        svm_classifier.fit(features, y_train)
        
        # Calculate training time
        train_time = time.time() - start_time
        model_train_times[f'{model_name}_svm'] = train_time
        
        # Training metrics
        y_train_pred = svm_classifier.predict(features)
        train_accuracy = accuracy_score(y_train, y_train_pred)
        
        # Validation metrics
        val_features = scaled_val_features[model_name]
        y_val_pred = svm_classifier.predict(val_features)
        val_accuracy = accuracy_score(y_val, y_val_pred)
        
        log_message(f"{model_name} SVM Training Time: {train_time:.2f} seconds")
        log_message(f"{model_name} SVM Training Accuracy: {train_accuracy:.4f}")
        log_message(f"{model_name} SVM Validation Accuracy: {val_accuracy:.4f}")
        
        # Save model and scaler
        joblib.dump(svm_classifier, f"/kaggle/working/{model_name}_svm_multi_ham10000.joblib")
        joblib.dump(scalers[model_name], f"/kaggle/working/{model_name}_scaler_svm_multi_ham10000.joblib")
        
        # Store results
        results[f'{model_name}_svm'] = {
            'model': svm_classifier,
            'scaler': scalers[model_name],
            'train_acc': train_accuracy,
            'val_acc': val_accuracy
        }
        
        # Train Random Forest
        log_message(f"\n{'='*50}")
        log_message(f"Training Random Forest using {model_name} features...")
        
        check_progress_update(f"Training Random Forest with {model_name} features")
        
        # Record start time
        start_time = time.time()
        
        rf_classifier = RandomForestClassifier(**RF_PARAMS)
        rf_classifier.fit(features, y_train)
        
        # Calculate training time
        train_time = time.time() - start_time
        model_train_times[f'{model_name}_rf'] = train_time
        
        # Training metrics
        y_train_pred = rf_classifier.predict(features)
        train_accuracy = accuracy_score(y_train, y_train_pred)
        
        # Validation metrics
        y_val_pred = rf_classifier.predict(val_features)
        val_accuracy = accuracy_score(y_val, y_val_pred)
        
        log_message(f"{model_name} RF Training Time: {train_time:.2f} seconds")
        log_message(f"{model_name} RF Training Accuracy: {train_accuracy:.4f}")
        log_message(f"{model_name} RF Validation Accuracy: {val_accuracy:.4f}")
        
        # Save model
        joblib.dump(rf_classifier, f"/kaggle/working/{model_name}_rf_multi_ham10000.joblib")
        joblib.dump(scalers[model_name], f"/kaggle/working/{model_name}_scaler_rf_multi_ham10000.joblib")
        
        # Store results
        results[f'{model_name}_rf'] = {
            'model': rf_classifier,
            'scaler': scalers[model_name],
            'train_acc': train_accuracy,
            'val_acc': val_accuracy
        }
        
        # Release memory 
        release_memory()
    
    return results

def train_fusion_models(X_train_fused, y_train, X_val_fused, y_val):
    """Train models on fused features from all feature extractors"""
    log_message("\n" + "="*50)
    log_message("Training SVM classifier on fused features...")
    
    check_progress_update("Training SVM on fused features")
    
    # Scale fused features
    log_message("Scaling fused features...")
    scaler = StandardScaler()
    X_train_fused_scaled = scaler.fit_transform(X_train_fused)
    X_val_fused_scaled = scaler.transform(X_val_fused)
    
    # Save scaler
    joblib.dump(scaler, "/kaggle/working/fusion_scaler_svm_multi_ham10000.joblib")
    
    # Record start time
    start_time = time.time()
    
    #  SVM parameters 
    svm_classifier = SVC(
        kernel='rbf',           
        probability=True,
        C=1.0,  
        gamma='scale' ,
        cache_size=3000,          
        max_iter=400,         
        tol=1e-2,              
        class_weight=None , 
        verbose=False,  
        decision_function_shape='ovr',
        break_ties= False   
    )

     
    svm_classifier.fit(X_train_fused_scaled, y_train)
    
    # Calculate training time
    train_time = time.time() - start_time
    model_train_times['fusion_svm'] = train_time
    log_message(f"Fusion SVM training time: {train_time:.2f} seconds")
    
    # Training metrics
    y_train_pred = svm_classifier.predict(X_train_fused_scaled)
    train_accuracy = accuracy_score(y_train, y_train_pred)
    
    # Validation metrics
    y_val_pred = svm_classifier.predict(X_val_fused_scaled)
    val_accuracy = accuracy_score(y_val, y_val_pred)
    
    log_message(f"Fusion SVM Training Accuracy: {train_accuracy:.4f}")
    log_message(f"Fusion SVM Validation Accuracy: {val_accuracy:.4f}")
    
    # Save the model
    joblib.dump(svm_classifier, model_paths['svm'])
    log_message(f"Fusion SVM model saved to {model_paths['svm']}")
    
    # Train Random Forest
    log_message("\n" + "="*50)
    log_message("Training Random Forest classifier on fused features...")
    
    check_progress_update("Training Random Forest on fused features")
    
    # Record start time
    start_time = time.time()
    
    #  Random Forest 
    rf_classifier = RandomForestClassifier(
        n_estimators=100,
        random_state=42,
        n_jobs=-1, 
        max_depth=10,      
        min_samples_split=10,   
        min_samples_leaf=4,    
        bootstrap=True,    
        verbose=0,
        warm_start=False,
        max_features='sqrt',  
        criterion= 'gini',     
        max_leaf_nodes= 100     
    )
     
    rf_classifier.fit(X_train_fused_scaled, y_train)
    
    # Calculate training time
    train_time = time.time() - start_time
    model_train_times['fusion_rf'] = train_time
    log_message(f"Fusion Random Forest training time: {train_time:.2f} seconds")
    
    # Training metrics
    y_train_pred = rf_classifier.predict(X_train_fused_scaled)
    train_accuracy = accuracy_score(y_train, y_train_pred)
    
    # Validation metrics
    y_val_pred = rf_classifier.predict(X_val_fused_scaled)
    val_accuracy = accuracy_score(y_val, y_val_pred)
    
    log_message(f"Fusion Random Forest Training Accuracy: {train_accuracy:.4f}")
    log_message(f"Fusion Random Forest Validation Accuracy: {val_accuracy:.4f}")
    
    # Save the model and scaler
    joblib.dump(rf_classifier, model_paths['random_forest'])
    joblib.dump(scaler, "/kaggle/working/fusion_scaler_rf_multi_ham10000.joblib")
    log_message(f"Fusion Random Forest model saved to {model_paths['random_forest']}")
    
    return {
        'fusion_svm': {
            'model': svm_classifier,
            'scaler': scaler,
            'train_acc': train_accuracy,
            'val_acc': val_accuracy
        },
        'fusion_rf': {
            'model': rf_classifier,
            'scaler': scaler,
            'train_acc': train_accuracy,
            'val_acc': val_accuracy
        }
    }

def print_detailed_metrics(y_true, y_pred, y_prob, idx_to_class, model_name):
    """Print detailed metrics for a model including confusion matrix and classification report"""
    num_classes = len(idx_to_class)
    class_names = [idx_to_class[i] for i in range(num_classes)]
    
    # Generate classification report
    report = classification_report(y_true, y_pred, target_names=class_names, output_dict=True)
    
    # Extract precision, recall, and F1-score
    precision = {}
    recall = {}
    f1_score = {}
    support = {}
    
    for i, class_name in enumerate(class_names):
        precision[class_name] = report[class_name]['precision']
        recall[class_name] = report[class_name]['recall']
        f1_score[class_name] = report[class_name]['f1-score']
        support[class_name] = report[class_name]['support']
    
    # Create a DataFrame 
    metrics_df = pd.DataFrame({
        'Class': class_names,
        'Precision': [precision[cls] for cls in class_names],
        'Recall': [recall[cls] for cls in class_names],
        'F1-Score': [f1_score[cls] for cls in class_names],
        'Support': [support[cls] for cls in class_names]
    })
    
    # Print the detailed metrics
    log_message(f"\n{'-'*30} {model_name} Metrics {'-'*30}")
    log_message(f"Overall Accuracy: {report['accuracy']:.4f}")
    log_message(f"\nDetailed Metrics by Class:")
    log_message(metrics_df.to_string(index=False))
    
    # Print macro and weighted averages
    log_message(f"\nMacro Average:")
    log_message(f"  Precision: {report['macro avg']['precision']:.4f}")
    log_message(f"  Recall: {report['macro avg']['recall']:.4f}")
    log_message(f"  F1-Score: {report['macro avg']['f1-score']:.4f}")
    
    log_message(f"\nWeighted Average:")
    log_message(f"  Precision: {report['weighted avg']['precision']:.4f}")
    log_message(f"  Recall: {report['weighted avg']['recall']:.4f}")
    log_message(f"  F1-Score: {report['weighted avg']['f1-score']:.4f}")
    
    # Save metrics to global dictionary
    all_metrics[model_name] = {
        'accuracy': report['accuracy'],
        'precision': precision,
        'recall': recall,
        'f1_score': f1_score,
        'macro_avg_precision': report['macro avg']['precision'],
        'macro_avg_recall': report['macro avg']['recall'],
        'macro_avg_f1': report['macro avg']['f1-score'],
        'weighted_avg_precision': report['weighted avg']['precision'],
        'weighted_avg_recall': report['weighted avg']['recall'],
        'weighted_avg_f1': report['weighted avg']['f1-score']
    }
    
    # Plot confusion matrix
    cm = confusion_matrix(y_true, y_pred)
    
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=class_names, 
                yticklabels=class_names)
    plt.xlabel('Predicted Label', fontsize=14)
    plt.ylabel('True Label', fontsize=14)
    plt.title(f'Confusion Matrix - {model_name}', fontsize=16)
    
    plt.subplots_adjust(bottom=0.15, left=0.15, right=0.95, top=0.9)
    plt.savefig(f"/kaggle/working/{model_name}_confusion_matrix_multi_ham10000.png", dpi=300)
    plt.show()

def plot_roc_curves(y_test, y_test_prob, num_classes, idx_to_class, model_name):
    """Plot ROC curves for a model"""
    
    y_test_bin = label_binarize(y_test, classes=range(num_classes))
    
    # Compute micro-average ROC curve and ROC area
    fpr_micro, tpr_micro, _ = roc_curve(y_test_bin.ravel(), y_test_prob.ravel())
    roc_auc_micro = auc(fpr_micro, tpr_micro)
    
    # Compute ROC curve and ROC area for each class
    fpr = {}
    tpr = {}
    roc_auc = {}
    
    for i in range(num_classes):
        fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_test_prob[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])
    
    # Plot ROC curve
    plt.figure(figsize=(12, 10))
    
    # Plot ROC curves for each class
    colors = ['aqua', 'darkorange', 'cornflowerblue', 'green', 'red', 'purple', 'brown']
    label_names = [idx_to_class[i] for i in range(num_classes)]
    
    for i, color in zip(range(min(num_classes, len(colors))), colors):
        plt.plot(fpr[i], tpr[i], color=color, lw=2,
                 label=f'{label_names[i]} (AUC = {roc_auc[i]:.4f})')
    
    # Plot micro-average ROC curve
    plt.plot(fpr_micro, tpr_micro,
             label=f'Micro-average (AUC = {roc_auc_micro:.4f})',
             color='deeppink', linestyle=':', linewidth=4)
    
    # Add diagonal line
    plt.plot([0, 1], [0, 1], 'k--', lw=2)
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate', fontsize=14)
    plt.ylabel('True Positive Rate', fontsize=14)
    plt.title(f'ROC Curve for {model_name} (Multi Classification)', fontsize=16)
    plt.legend(loc="lower right", fontsize=12)
    plt.grid(alpha=0.3)
    
    plt.subplots_adjust(bottom=0.1, left=0.1, right=0.9, top=0.9)
    plt.savefig(f"/kaggle/working/{model_name}_roc_curve_multi_ham10000.png", dpi=300)
    plt.show()

def evaluate_individual_models(models_dict, X_test_dict, y_test, idx_to_class, num_classes):
    """Evaluate individual models and print detailed metrics"""
    results = {}
    
    for model_key, model_info in models_dict.items():
        if model_key.startswith('fusion'):
            continue  # Skip fusion models for now
            
        # Extract model name and type
        model_parts = model_key.split('_')
        feature_type = model_parts[0]
        classifier_type = model_parts[1]
        
        log_message(f"\nEvaluating {classifier_type.upper()} with {feature_type} features...")
        
        check_progress_update(f"Evaluating {classifier_type.upper()} with {feature_type} features")
        
        # Record start time
        start_time = time.time()
        
        # Get the model, scaler and features
        model = model_info['model']
        scaler = model_info['scaler']
        X_test_scaled = scaler.transform(X_test_dict[feature_type])
        
        # Make predictions
        y_test_pred = model.predict(X_test_scaled)
        test_accuracy = accuracy_score(y_test, y_test_pred)
        
        # Get probabilities
        y_test_prob = model.predict_proba(X_test_scaled)
        
        # Calculate evaluation time
        eval_time = time.time() - start_time
        model_eval_times[model_key] = eval_time
        model_total_times[model_key] = model_train_times[model_key] + eval_time
        
        log_message(f"{feature_type} {classifier_type.upper()} Test Accuracy: {test_accuracy:.4f}")
        log_message(f"Evaluation Time: {eval_time:.2f} seconds")
        
        # Print metrics and plot confusion matrix
        print_detailed_metrics(y_test, y_test_pred, y_test_prob, idx_to_class, model_key)
        
        # Plot ROC curves
        plot_roc_curves(y_test, y_test_prob, num_classes, idx_to_class, model_key)
        
        # Store results
        results[model_key] = {
            'test_acc': test_accuracy,
            'eval_time': eval_time,
            'predictions': y_test_pred,
            'probabilities': y_test_prob
        }
    
    return results

def evaluate_fusion_models(models_dict, X_test_fused, y_test, idx_to_class, num_classes):
    """Evaluate fusion models and print detailed metrics"""
    results = {}
    
    for model_key in ['fusion_svm', 'fusion_rf']:
        model_info = models_dict[model_key]
        classifier_type = model_key.split('_')[1].upper()
        
        log_message(f"\nEvaluating Fusion {classifier_type}...")
        
        check_progress_update(f"Evaluating Fusion {classifier_type}")
        
        # Record start time
        start_time = time.time()
        
        # Get the model and scaler
        model = model_info['model']
        scaler = model_info['scaler']
        X_test_fused_scaled = scaler.transform(X_test_fused)
        
        # Make predictions
        y_test_pred = model.predict(X_test_fused_scaled)
        test_accuracy = accuracy_score(y_test, y_test_pred)
        
        # Get probabilities
        y_test_prob = model.predict_proba(X_test_fused_scaled)
        
        # Calculate evaluation time
        eval_time = time.time() - start_time
        model_eval_times[model_key] = eval_time
        model_total_times[model_key] = model_train_times[model_key] + eval_time
        
        log_message(f"Fusion {classifier_type} Test Accuracy: {test_accuracy:.4f}")
        log_message(f"Evaluation Time: {eval_time:.2f} seconds")
        
        # Print metrics and plot confusion matrix
        print_detailed_metrics(y_test, y_test_pred, y_test_prob, idx_to_class, model_key)
        
        # Plot ROC curves
        plot_roc_curves(y_test, y_test_prob, num_classes, idx_to_class, model_key)
        
        # Store results
        results[model_key] = {
            'test_acc': test_accuracy,
            'eval_time': eval_time,
            'predictions': y_test_pred,
            'probabilities': y_test_prob
        }
    
    return results

def compare_all_models(all_results):
    """Compare performance of all models"""
    # Collect test accuracies
    model_names = []
    test_accuracies = []
    
    for model_key, results in all_results.items():
        model_names.append(model_key)
        test_accuracies.append(results['test_acc'])
    
    # Create dataframe 
    comparison_df = pd.DataFrame({
        'Model': model_names,
        'Test Accuracy': test_accuracies
    })
    
    # Add more metrics from the all_metrics dictionary
    comparison_df['Macro Avg F1'] = [all_metrics[model]['macro_avg_f1'] for model in model_names]
    comparison_df['Training Time (s)'] = [model_train_times.get(model, 0) for model in model_names]
    comparison_df['Evaluation Time (s)'] = [model_eval_times.get(model, 0) for model in model_names]
    comparison_df['Total Time (s)'] = [model_total_times.get(model, 0) for model in model_names]
    
    # Sort by accuracy
    comparison_df = comparison_df.sort_values(by='Test Accuracy', ascending=False)
    
    log_message("\nModel Performance Comparison:")
    log_message(comparison_df.to_string(index=False))
    
    # Save comparison 
    comparison_df.to_csv('/kaggle/working/model_comparison_multi_ham10000.csv', index=False)
    
    # Plot comparison chart for accuracy
    plt.figure(figsize=(14, 8))
    
    # Create a colormap 
    colors = ['lightgreen' if 'fusion' in name else 'skyblue' for name in comparison_df['Model']]
    
    # Create the bars with labels
    bars = plt.bar(comparison_df['Model'], comparison_df['Test Accuracy'], color=colors)
    
    # Add values on top of bars
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height + 0.005,
                f'{height:.4f}', ha='center', va='bottom', fontweight='bold')
    
    plt.xlabel('Models', fontsize=14)
    plt.ylabel('Test Accuracy', fontsize=14)
    plt.title('Test Accuracy Comparison: Multi-Class Classification (7 Classes)', fontsize=16)
    plt.xticks(rotation=45, ha='right')
    plt.ylim(0, max(test_accuracies) + 0.05)
    plt.grid(axis='y', linestyle='--', alpha=0.3)
    
    # Create proper legend
    fusion_patch = plt.Rectangle((0,0),1,1, color='lightgreen', label='Fusion Models')
    individual_patch = plt.Rectangle((0,0),1,1, color='skyblue', label='Individual Models')
    plt.legend(handles=[fusion_patch, individual_patch])
    
    plt.subplots_adjust(bottom=0.25, left=0.1, right=0.95, top=0.9)
    plt.savefig('/kaggle/working/all_models_accuracy_comparison_multi_ham10000.png', dpi=300)
    plt.show()
    
    return comparison_df




# Define the main function 
def main():
    # Initialize progress update time
    global last_progress_update
    last_progress_update = time.time()
    
    # Set device for feature extraction
    device, device_type = setup_device()
    log_message(f"Using device: {device}")
    
    # Apply optimizations to PyTorch
    if device_type == 'cuda':
        torch.backends.cudnn.benchmark = True
        torch.backends.cuda.matmul.allow_tf32 = True
        torch.backends.cudnn.allow_tf32 = True
    elif device_type == 'tpu':
        log_message("Using TPU-specific optimizations")
    
    # Dataset path
    dataset_path = "/kaggle/input/skin-cancer-mnist-ham10000/"
    
    # Run augmentation and preparation 
    log_message("Starting dataset preparation...")
    dataset_info = augment_and_prepare_dataset(dataset_path)
    
    # Extract features for training set
    log_message("\nExtracting features from training set...")
    X_train_dict, X_train_fused, y_train = extract_features_and_labels(dataset_info['train_loader'], device, device_type)
    
    # Extract features for validation set
    log_message("\nExtracting features from validation set...")
    X_val_dict, X_val_fused, y_val = extract_features_and_labels(dataset_info['val_loader'], device, device_type)
    
    # Extract features for test set
    log_message("\nExtracting features from test set...")
    X_test_dict, X_test_fused, y_test = extract_features_and_labels(dataset_info['test_loader'], device, device_type)
    
    # Get number of classes
    num_classes = dataset_info['num_classes']
    log_message(f"\nNumber of classes: {num_classes}")
    assert num_classes == 7, "Expected 7 classes for multi-class classification"
    
    # Train individual models for each feature extractor
    log_message("\nTraining individual models...")
    individual_models = train_individual_models(X_train_dict, y_train, y_val, X_val_dict)
    
    # Train models on fused features
    log_message("\nTraining fusion models...")
    fusion_models = train_fusion_models(X_train_fused, y_train, X_val_fused, y_val)
    
    # Combine all models for evaluation
    all_models = {**individual_models, **fusion_models}
    
    # Evaluate individual models
    log_message("\nEvaluating individual models...")
    individual_results = evaluate_individual_models(all_models, X_test_dict, y_test, dataset_info['idx_to_class'], num_classes)
    
    # Evaluate fusion models
    log_message("\nEvaluating fusion models...")
    fusion_results = evaluate_fusion_models(all_models, X_test_fused, y_test, dataset_info['idx_to_class'], num_classes)
    
    # Combine all results
    all_results = {**individual_results, **fusion_results}
    
    # Compare all models
    log_message("\nComparing all models...")
    comparison = compare_all_models(all_results)
 
    
   
    return comparison

if __name__ == "__main__":
    # Fix for SSL certificate issue
    ssl._create_default_https_context = ssl._create_unverified_context

    # For 10-minute progress updates
    last_progress_update = time.time()

    def check_progress_update(message):
        """Check if 10 minutes have passed to update progress"""
        global last_progress_update
        current_time = time.time()
        
        if current_time - last_progress_update >= 600:  # 10 minutes = 600 seconds
            log_message(f"Progress update: {message}")
            last_progress_update = current_time
            return True
        return False

    # Determine optimal thread count based on system
    cpu_count = multiprocessing.cpu_count()
    optimal_threads = max(1, cpu_count - 1)  

    # Set environment variables for optimized performance
    os.environ['OMP_NUM_THREADS'] = str(optimal_threads)
    os.environ['MKL_NUM_THREADS'] = str(optimal_threads)
    os.environ['NUMEXPR_NUM_THREADS'] = str(optimal_threads)
    os.environ['VECLIB_MAXIMUM_THREADS'] = str(optimal_threads)
    os.environ['OPENBLAS_NUM_THREADS'] = str(optimal_threads)
    
    # Dictionary to store model paths
    model_paths = {
        'svm': "/kaggle/working/feature_fusion_svm_multi_ham10000.joblib",
        'random_forest': "/kaggle/working/feature_fusion_rf_multi_ham10000.joblib"
    }

    # Dictionaries to store metrics
    model_train_times = {}
    model_eval_times = {}
    model_total_times = {}
    all_metrics = {}
    
    # Set random seeds for reproducibility
    torch.manual_seed(42)
    random.seed(42)
    np.random.seed(42)
    
    try:
        main()
    except Exception as e:
        log_message(f"ERROR: {str(e)}")
        # Save stack trace to log file
        import traceback
        log_message(traceback.format_exc())
```
